```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
ingreso_total = tienda['Precio'].sum()
ingreso_total2 = tienda2['Precio'].sum()
ingreso_total3 = tienda3['Precio'].sum()
ingreso_total4 = tienda4['Precio'].sum()
```

```python
print(ingreso_total)
print(ingreso_total2)
print(ingreso_total3)
print(ingreso_total4)
```

```python
import matplotlib.pyplot as plt

# Datos consolidados
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total, ingreso_total2, ingreso_total3, ingreso_total4]

plt.bar(tiendas, ingresos, color=['#4CAF50', '#2196F3', '#FF9800', '#F44336'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')
plt.savefig('1_ingresos_totales.png')
plt.show()
```

```python
# Contar productos por categoría en cada tienda
ventas_cat = tienda['Categoría del Producto'].value_counts()
ventas_cat2 = tienda2['Categoría del Producto'].value_counts()
ventas_cat3 = tienda3['Categoría del Producto'].value_counts()
ventas_cat4 = tienda4['Categoría del Producto'].value_counts()

# Mostrar las categorías más populares de la Tienda 1 como ejemplo
print("Categorías más vendidas - Tienda 1:")
print(ventas_cat)
print("\nCategorías más vendidas - Tienda 2:")
print(ventas_cat2)
print("\nCategorías más vendidas - Tienda 3:")
print(ventas_cat3)
print("\nCategorías más vendidas - Tienda 4:")
print(ventas_cat4)
```

```python

plt.figure(figsize=(8, 8))
ventas_cat.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='Pastel1')
plt.title('Distribución de Categorías - Tienda 1')
plt.ylabel('') # Quitar etiqueta lateral
plt.savefig('2_categorias_tienda1.png')
plt.show()
```

```python
media_calificacion = tienda['Calificación'].mean()
print("Media de la Tienda 1:")
print(media_calificacion)

media_calificacion2 = tienda2['Calificación'].mean()
print("\nMedia de la Tienda 2:")
print(media_calificacion2)

media_calificacion3 = tienda3['Calificación'].mean()
print("\nMedia de la Tienda 3:")
print(media_calificacion3)

media_calificacion4 = tienda4['Calificación'].mean()
print("\nMedia de la Tienda 4:")
print(media_calificacion4)
```

```python
medias = [media_calificacion, media_calificacion2, media_calificacion3, media_calificacion4]

plt.barh(tiendas, medias, color='skyblue')
plt.xlabel('Calificación Promedio (1-5)')
plt.title('Satisfacción del Cliente por Tienda')
plt.xlim(0, 5) # La escala de estrellas es hasta 5
plt.legend()
plt.savefig('3_satisfaccion_cliente.png')
plt.show()
```

```python
articulo_mas_vendido = tienda['Producto'].value_counts().idxmax()
print("Artículo más vendido - Tienda 1:")
print(articulo_mas_vendido)

articulo_menos_vendido = tienda['Producto'].value_counts().idxmin()
print("\nArtículo menos vendido - Tienda 1:")
print(articulo_menos_vendido)

articulo_mas_vendido2 = tienda2['Producto'].value_counts().idxmax()
print("\nArtículo más vendido - Tienda 2:")
print(articulo_mas_vendido2)

articulo_menos_vendido2 = tienda2['Producto'].value_counts().idxmin()
print("\nArtículo menos vendido - Tienda 2:")
print(articulo_menos_vendido2)

articulo_mas_vendido3 = tienda3['Producto'].value_counts().idxmax()
print("\nArtículo más vendido - Tienda 3:")
print(articulo_mas_vendido3)

articulo_menos_vendido3 = tienda3['Producto'].value_counts().idxmin()
print("\nArtículo menos vendido - Tienda 3:")
print(articulo_menos_vendido3)

articulo_mas_vendido4 = tienda4['Producto'].value_counts().idxmax()
print("\nArtículo más vendido - Tienda 4:")
print(articulo_mas_vendido4)

articulo_menos_vendido4 = tienda4['Producto'].value_counts().idxmin()
print("\nArtículo menos vendido - Tienda 4:")
print(articulo_menos_vendido4)
```

```python
promedio_envio = tienda['Costo de envío'].mean()
print("Promedio de costo de envío - Tienda 1:")
print(promedio_envio)

promedio_envio = tienda2['Costo de envío'].mean()
print("\nPromedio de costo de envío - Tienda 2:")
print(promedio_envio)

promedio_envio = tienda3['Costo de envío'].mean()
print("\nPromedio de costo de envío - Tienda 3:")
print(promedio_envio)

promedio_envio = tienda4['Costo de envío'].mean()
print("\nPromedio de costo de envío - Tienda 4:")
print(promedio_envio)
```

```python
envios = [tienda['Costo de envío'].mean(), tienda2['Costo de envío'].mean(),
          tienda3['Costo de envío'].mean(), tienda4['Costo de envío'].mean()]

plt.scatter(envios, ingresos, s=200, c=['green', 'blue', 'orange', 'red'], alpha=0.7)

# Anotar los puntos
for i, txt in enumerate(tiendas):
    plt.annotate(txt, (envios[i], ingresos[i]), xytext=(5,5), textcoords='offset points')

plt.title('Relación: Costo de Envío vs Ingresos')
plt.xlabel('Costo de Envío Promedio')
plt.ylabel('Ingresos Totales')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig('4_envio_vs_ingresos.png')
plt.show()
```
